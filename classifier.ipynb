#### Encoding

```python
symptoms = pd.read_csv('symptoms.csv')
symptoms.iloc[-1] = ['NaN', 1]
symptoms.head()
```

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
le.fit(symptoms['Symptom'])
```

#### xy split

```python
from sklearn.utils import shuffle
import numpy as np

df_new = shuffle(df_new)
df_new = df_new.apply(lambda x: x.str.replace(' ', ''))
df_new = df_new.fillna('NaN')
x = df_new.drop('Disease', axis=1)
y = df_new['Disease']

print(x.shape, y.shape)
```

```python
y.unique()
```

```python
# split data into train and test sets
from sklearn.model_selection import train_test_split, cross_val_score
np.random.seed(42)

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2)
```

#### Encode transformation

```python
X_train = X_train.apply(le.transform)
X_test = X_test.apply(le.transform)
```

#### Classifier evaluation

```python
from sklearn.naive_bayes import CategoricalNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Instantiate the classfiers and make a list
classifiers = [
    CategoricalNB(), 
    KNeighborsClassifier(), 
]
```

```python
from sklearn.metrics import accuracy_score
# Train the models and record the results
for cls in classifiers:
    model = cls.fit(X_train, Y_train)
    pred = cls.predict(X_test)
    
    score = accuracy_score(Y_test, pred)
    print('Accuracy of', cls, ':', score)
```

#### Final model

```python
classifier = CategoricalNB()

# completet dataset
x = x.apply(le.transform)

classifier.fit(x, y)
```

#### save model

```python
import pickle

file_name = 'disease_classifier.sav'

with open(file_name, 'wb') as f:
    pickle.dump(classifier, f)

```

```python
# load saved file                                                                                                                                                                                                           

clf = pickle.load(open(file_name, 'rb'))
```

```python
data = ['high_fever', 'nodal_skin_eruptions', 'stomach_pain', 'continuous_sneezing', 'NaN', 'cough']
```
